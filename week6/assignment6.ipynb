```python
import pandas as pd
```

```python
file_url = 'https://raw.githubusercontent.com/'\
'PacktWorkshops/'\
'The-Applied-Artificial-Intelligence-Workshop/'\
'master/Datasets/car.csv'
```

```python
df = pd.read_csv(file_url)
```

```python
df.head()
```

```python
from sklearn import preprocessing
```

```python
def encode(data_frame, column):
    """ This function will instantiate LabelEncoder(), fit it
with the unique value of the column, and transform its data. It will return the
transformed column. Given in the textbook """
    label_encoder = preprocessing.LabelEncoder()
    label_encoder.fit(data_frame[column].unique())
    return label_encoder.transform(data_frame[column])
```

```python
# A for loop to iterate on dataframe columns and run the encode function
for column in df.columns:
    df[column] = encode(df, column)
```

```python
df.head()
```

```python
# Taking out label out from the data frame
label = df.pop('class')
```

```python
label
```

```python
from sklearn import model_selection
```

```python
# Created the training and test and split them by 10%, random state
features_train, features_test, label_train, label_test = \
model_selection.train_test_split(df, label, \
test_size=0.1, \
random_state=42)
```

```python
features_train, features_test, label_train, label_test
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
# Initiate the decision tree class
decision_tree = DecisionTreeClassifier()
```

```python
# Running the decision treee fit function
decision_tree.fit(features_train, label_train)
```

```python
# Score on the featuer test and label test
decision_tree.score( features_test, label_test )
```

```python
from sklearn.metrics import classification_report
```

```python
(classification_report(label_test, \
decision_tree.predict(features_test)))
```

```python
# Ran the classification report method with argument of decsion_tree predict
print(classification_report(label_test, \
decision_tree.predict(features_test)))
```
