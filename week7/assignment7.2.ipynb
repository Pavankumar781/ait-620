```python
import pandas as pd
```

```python
file_url = "https://raw.githubusercontent.com/PacktWorkshops/The-Applied-Artificial-Intelligence-Workshop/refs/heads/master/Datasets/Sales_Transactions_Dataset_Weekly.csv"
```

```python
df = pd.read_csv(file_url)
```

```python
df.head()
```

```python
# Drop the first 55 columns
cols_55 = df.columns[:55]
```

```python
new_df = df.drop(columns=cols_55)
```

```python
new_df.head()
```

```python
from sklearn.cluster import KMeans
```

```python
# Created the k meand model with number of cluster 7 with random state 8
k_means_model = KMeans(n_clusters=8,random_state=8)
```

```python
# Ran the fit model method on data set
k_means_model.fit(new_df)
```

```python
centers = k_means_model.cluster_centers_
```

```python
centers
```

```python
# Get the label from our data set as k_means does from thw raw data
labels = k_means_model.labels_
```

```python
labels
```

```python
# Requirment of the activity and got the first 55 column now keep it 52 after removing min, max and product code
w_cols = df.columns[52:]
```

```python
df.drop(columns=w_cols, inplace=True)
```

```python
df.head()
```

```python
# Added the labels colum to data set
df['label'] = labels
```

```python
df.head()
```

```python
# Ran the aggegation as want by exercise and sum it
df_ag_lb = df.groupby('label').sum()
```

```python
df_ag_lb
```

```python
# Counted the labels and W0
df_result = df[['label','W0']].groupby('label').count()
```

```python
df_result
```

```python
# Rename the W0 as count_product
df_result= df_result.rename(columns = {'W0':'count_product'})
```

```python
df_result
```

```python
# Get the tolal sales of each row with label
df_result['total_sales'] = df_ag_lb.sum(axis = 1)
```

```python
df_result
```

```python
# Getting average sale by total sale with count product
df_result['yearly_average_sales']= df_result['total_sales'] / df_result['count_product']
```

```python
df_result
```

```python
# Sorted, ran the ascending and saved the final output
df_result.sort_values(by='yearly_average_sales', ascending=False, inplace = True)
```
